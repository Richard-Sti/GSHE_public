```julia
import BenchmarkTools: @btime, @benchmark;

import Pkg: activate, build
activate("../GSHEIntegrator/.")
import GSHEIntegrator;
```

```julia
geometry = GSHEIntegrator.setup_geometry(rsource=3, θsource= π/2, ϕsource=0, robs=50, θobs=π/3, ϕobs=π, a=1.0, ϵ=0.0,);
```

Initial $p_i$ covector:

```julia
println(@btime GSHEIntegrator.pi0(0.2, 0.3, geometry))
println(@code_warntype GSHEIntegrator.pi0(0.2, 0.3, geometry))
```

Initial values: $(t, r, \theta, \phi, p_r, p_\theta, p_\phi)$

```julia
p = [0.2, 0.3]
x0 = GSHEIntegrator.init_values(p, geometry)
println(@btime GSHEIntegrator.init_values(p, geometry))
println(@code_warntype GSHEIntegrator.init_values(p, geometry))
```

Null condition: $p_t$

```julia
println(@btime GSHEIntegrator.pt_null(x0, geometry.params.a))
println(@code_warntype GSHEIntegrator.pt_null(x0, geometry.params.a))
```

Tetrad boosting

```julia
println(@btime GSHEIntegrator.tetrad_boosting(4, geometry))
println(@code_warntype GSHEIntegrator.tetrad_boosting(4, geometry))
```

```julia
println(@btime GSHEIntegrator.derivative_tetrad_boosting(4, geometry))
println(@code_warntype GSHEIntegrator.derivative_tetrad_boosting(4, geometry))
```
