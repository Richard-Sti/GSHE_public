```python
import numpy as np

%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib as mpl
import healpy as hp

import sys
sys.path.append("../")
import GSHEWaveform
```

```python
N = 50
k2 = np.linspace(-1, 1, N)
k3 = np.linspace(-1, 1, N)

k2, k3 = [x.reshape(-1, ) for x in np.meshgrid(k2, k3)]


m = (k2**2 + k3**2) <= 1
psi = np.arccos(k3)
rho = np.pi + np.arcsin(k2 / np.sqrt(1 - k3**2))

dist = np.asarray([hp.rotator.angdist([psi[i], rho[i]], [np.pi/2, np.pi])[0] for i in range(N**2)])
```

```python
GSHEWaveform.setmplstyle("../plot_style.txt")
# GSHEWaveform.setmplstyle()

plt.figure()

plt.scatter(rho[m], psi[m], c=dist[m])

plt.xlabel(r"$\rho$")
plt.ylabel(r"$\psi$")



plt.savefig("/Users/richard/Downloads/shadow_coords.pdf", bbox_inches="tight", dpi=300)

plt.show()
```
