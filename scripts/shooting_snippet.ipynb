```python
import numpy as np

%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.colors as col

import joblib
```

```python
data = joblib.load("../../notebooks/gshe_data/data.p")
```

```python
"../results/"
```

```python
X, Y, Z = (data[p] for p in ("X", "Y", "Z"))
#time delay in ms, eq 3.6a
T=15*10**Z
```

```python
qs = [1, 99.9] # Colourbar percentiles

fig, ax = plt.subplots()

vmin, vmax = np.percentile(T[~np.isnan(T)].reshape(-1,), qs)
cm = ax.pcolormesh(X, Y, T, shading="auto", cmap="nipy_spectral", rasterized=True, vmin=vmin, vmax=vmax,norm=col.LogNorm())


cbar = fig.colorbar(cm, ax=ax)#
ax.set_title(r"Time delay [ms] $(\frac{10^4 M_\odot}{M})(\frac{40Hz}{f})^2$",fontsize=13)
#cbar.ax.tick_params(labelsize=14)

ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])
ax.axis('off')

#x.set_xlabel("$k_x$")
#x.set_ylabel("$k_y$")
ax.annotate('BH shadow',(-.05,0),fontsize=13)
plt.savefig("/home/miguel/code/GSHE/paper_plots/Delta_t.png", dpi=600, bbox_inches="tight")
fig.show()
```
